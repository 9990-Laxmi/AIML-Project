```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = pd.read_excel('/content/cleaned_water_consumption.csv.1.xlsx')
```

```python
print(data.head())
```

```python
# calculate mean,median,mode
mean = data['Total Water Consumption '].mean()
median = data['Total Water Consumption '].median()
mode = data['Total Water Consumption '].mode()
std = data['Total Water Consumption '].std()
```

```python
#print statistics
print(mean)
print(median)
print(mode)
print(std)
```

```python
#plot a histogram of water consumption
plt.figure(figsize=(10,6))
plt.hist(data['Total Water Consumption '])
```

```python
#plot scatter plot of water consumption over date
plt.figure(figsize=(10,6))
plt.scatter(data['Year'],data['Total Water Consumption '])
```
